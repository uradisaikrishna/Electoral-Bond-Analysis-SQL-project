> Looking into the row count of donor and reciver data we notice that they are not same that means may be there are donations no party has claime to donations no company has associated with. So lets combine these table and look into how that falls. 

- Prefix and BondNumber are few of the common columns in both the tabels and combining them together acts as the common and unique columns to identify the records and make a common connection between donors and recivers table. 

# Creating Unique Key (Primary Key for data)

- Combining mulitple columns to form a key is called as composite key which act as the primary key and the key with helps us in connecting  the receiver table and donor table. 

```python
donors_df['Unique_key'] = donors_df["Prefix"].astype(str) + donors_df["BondNumber"].astype(str)
donors_df.head()
```

```python
reciver_df['Unique_key'] = reciver_df['Prefix'].astype(str) + reciver_df['BondNumber'].astype(str)
reciver_df.head()
```

### Merging both the tables with a outer join: 

- Outer Join will give the combinations of donors and recivers data with nulls in missing data but overally you get some of the other detail corresponding every bond. 

```python
alldata_df = pd.merge(donors_df,reciver_df,how='outer',on='Unique_key')
alldata_df.head()
```

```python
# Finding the total records in combined data
alldata_df.shape
```

> The row count is way larger than the donors and recivers data so that means there are records in donor data which are not recived by any political party and records in reciver data no company has claimed.

```python
# Listing elements which don't have purchaser (Donor) and Party. 
allnulls_df = pd.DataFrame(alldata_df.isnull().sum())
allnulls_df.loc[['Purchaser','PartyName']]
```

> Above count shows that there are some bonds that dont have donor and some without political party.

> Taking Unique key and Denomination column will give total details of the bonds and we use this as a sepeate table in our database. 

```python
bonddata_df = alldata_df[['Unique_key','Denominations_x','Denominations_y']] 
bonddata_df
```

> In the merged data table, the columns `Denominations_x` and `Denominations_y` represent the denominations of bonds from the donor and receiver data, respectively. When both the donor and receiver names are present in our data, both columns will have the same value. However, in scenarios where only one of the columns has a value and the other is `NaN` or Null, replacing the null value with the value from the adjacent column will provide the value of each bond along with its unique number. Let's perform this replacement to obtain the value of each bond.

```python
bonddata_df['Denomination'] = bonddata_df['Denominations_x'].fillna(bonddata_df['Denominations_y'])
bonddata_df = bonddata_df[['Unique_key','Denomination']]
```
